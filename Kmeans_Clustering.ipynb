```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
data = load_iris()
X = data.data  # Features

# Optional: Standardize the features for better performance of K-means
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-means clustering
k = 3  # Number of clusters (for Iris dataset, we know there are 3 species)
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X_scaled)  # Fit the K-means model to the data


# Get the cluster labels (which data point belongs to which cluster)
labels = kmeans.labels_


# Get the centroids of the clusters
centroids = kmeans.cluster_centers_

# Plot the clusters (2D projection of the first two features for simplicity)
plt.figure(figsize=(8, 6))

# Plot each point with a color corresponding to its cluster label
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap='viridis', s=100, edgecolors='k', alpha=0.6)

# Plot the centroids
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=200, label='Centroids')

# Add labels and title
plt.xlabel("Feature 1 (Sepal Length)")
plt.ylabel("Feature 2 (Sepal Width)")
plt.title("K-means Clustering (Iris Dataset)")
plt.legend()
plt.show()
```
