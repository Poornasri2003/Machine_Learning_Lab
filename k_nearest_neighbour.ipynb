## Implementation of K Nearest Neighbor Algorithm


The K-Nearest Neighbors (KNN) algorithm is one of the simplest and most intuitive machine learning algorithms used for classification and regression tasks. It's a supervised learning algorithm that makes predictions based on the labels (or values) of the closest data points in the feature space.

#### How KNN Works:

1. **Input:**  You provide a dataset with known labels (for supervised learning). The algorithm stores the entire training dataset.

2. **Distance Calculation:**  When a new, unseen data point is presented (a test point), the algorithm calculates the distance between the test point and all the points in the training dataset. The most common distance metrics are:

Euclidean Distance: The straight-line distance between two points in Euclidean space.
Manhattan Distance: The sum of the absolute differences of the coordinates.
Cosine Similarity: Measures the cosine of the angle between two vectors (used in some variations).

3. **Find K Neighbors:** The algorithm then identifies the K nearest neighbors (the K training points with the smallest distances to the test point).

4. **Voting (Classification):** In classification tasks, the algorithm votes on the most frequent class among the K neighbors. The class with the highest number of votes is assigned to the test point.

For example, if we have K=5 and 3 neighbors belong to class A and 2 neighbors belong to class B, the test point will be classified as A.
Averaging (Regression): In regression tasks, the algorithm takes the average of the labels of the K nearest neighbors to predict a continuous value.

6. **Output:** After classifying the test point, the algorithm returns the predicted label (for classification) or value (for regression)

#### Importing necessary libraries

```python
import numpy as np
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
import seaborn as sns
```

```python
data=load_iris()
x=data.data
y=data.target
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.5)
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
y_prdeic=knn.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,y_prdeic)
print(accuracy)
```

```python
target_names = data.target_names
```

```python
import matplotlib.pyplot as plt
```

```python
cm = confusion_matrix(y_test, y_prdeic)

# Plot confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=data.target_names)
disp.plot(cmap='Blues')
plt.title("Confusion Matrix of KNN Model")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=y_test,y=y_prdeic)
```
